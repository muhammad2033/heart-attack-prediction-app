# train and test split

```python
from sklearn.model_selection import train_test_split 
```

```python
TrainData , TestData ,TrainTarget, testTarget = train_test_split(features, target , test_size = 0.20, random_state=0)
```

```python
TrainData = np.array(TrainData)
TestData = np.array(TestData)
TrainTarget = np.array(TrainTarget)
TestTarget = np.array(testTarget)
```

```python
TrainData.shape , TestData.shape , TrainTarget.shape , testTarget.shape
```

# fitting the ML model

```python
from sklearn.linear_model import LogisticRegression
```

# Logistic Regression


```python
classifier = LogisticRegression()
```

```python
classifier.fit(TrainData ,TrainTarget)
```

```python
# TestData[0] , testTarget[0] 
```

```python
testPreds = classifier.predict(TestData)
trainPreds =classifier.predict(TrainData)
```

```python
testPreds
```

```python
trainPreds
```

```python
# classifier.score(testTarget, testPreds)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

# # Assuming testPreds are probability scores for binary classification
# threshold = 0.5
# binary_preds = np.where(testPreds >= threshold, 1, 0)

TestAccuracy = round(accuracy_score(testTarget, testPreds)*100,2)
print("Test Accuracy is", TestAccuracy , "%")
```

```python
# # Assuming testPreds are probability scores for binary classification
# threshold = 0.5
# binary_preds = np.where(trainPreds >= threshold, 1, 0)

TrainAccuracy = round(accuracy_score(TrainTarget, trainPreds)*100,2)
print("Train Accuracy is", TrainAccuracy,"%")
```

# Decison Tree classifier

```python
#Model training
from sklearn.tree import DecisionTreeClassifier
```

```python
Dtree = DecisionTreeClassifier()
Dtree.fit(TrainData ,TrainTarget)
```

```python
# Evaluation 
from sklearn.metrics import accuracy_score
Pred_test = Dtree.predict(TestData)
print("Decision tree Evaluation")
accuracy = round(accuracy_score(TestTarget , Pred_test)*100,2)
print("accuray is ",accuracy ,"%")
```

```python
import pickle 
with open("Dtree", "wb") as new:
    pickle.dump(Dtree , new)
```
